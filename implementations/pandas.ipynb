```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'product_name': [
        'Product A', 'Product A', 'Product A',
        'Product B', 'Product B',
        'Product C', 'Product C',
        'Product D',
        'Product E', 'Product E',
    ],
    'rating': [4.5, 3.8, 4.2, 4.0, 3.5, 3.5, 4.1, 4.8, 3.2, 3.9],
    'review_text': [
        'Great product!', 'Decent but overpriced.', 'Solid quality.',
        'Good value for money.', 'Could be better.',
        'Average quality.', 'Surprisingly good!',
        'Excellent!',
        'Poor quality.', 'Improved over time.',
    ],
    "price": [100, 110, 95, 50, 55, 200, 180, 75, 15, 20],
})
```

```python
df.head()
```

```python
avg_rating = df[df["price"] > 50]["rating"].mean()
print(f"Average rating for products priced above 50: {avg_rating}")
```

```python
product_with_highest_rating_under_20 = df[df["price"] < 20]['rating'].idxmax()
print(product_with_highest_rating_under_20)
product_info = df.loc[product_with_highest_rating_under_20]
print(f"Product with the highest rating among those priced under 20: {product_info['product_name']} with a rating of {product_info['rating']}")
```

```python
avg_rating_product = df.groupby("product_name")["rating"].mean()
print("Average rating for each product:")
print(avg_rating_product)
```

```python
product_with_highest_rating = avg_rating_product.idxmax()
print(f"Product with the highest rating: {product_with_highest_rating}")
```

```python
no_reviews_per_product = df.groupby("product_name")["review_text"].count()
print("Number of reviews for each product:")
print(no_reviews_per_product)
```

```python
grouped_df = df.groupby("product_name").agg({
    "rating": "mean",
    "review_text": "count",
    "price": "mean"
})
grouped_df
```

```python
categories_df = pd.DataFrame({
    'product_name': ['Product A', 'Product B', 'Product C', 'Product D', 'Product E'],
    'category': ['Electronics', 'Clothing', 'Electronics', 'Home & Kitchen', 'Clothing']
})
```

```python
merged_df = df.merge(categories_df, on='product_name')
merged_df
```

```python
avg_rating_per_category = merged_df.groupby("category")["rating"].mean()
print("Average rating for each category:")
print(avg_rating_per_category)
```
