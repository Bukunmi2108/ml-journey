# Charts

```python
sns.histplot(data=df, x='Age', kde=True, color='skyblue')

plt.title('Histogram of Age: Checking Distribution Shape')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(data=df, x='Income (USD)', kde=True, color='skyblue')

plt.title('Histogram of Income (USD): Who earns what')
plt.xlabel('Income (USD)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Gender', palette='Set2')
plt.title('Gender Balance')
plt.show()
```

```python
sns.boxplot(data=df, x="Income (USD)", color='lightblue')
plt.title('Boxplot of Income (USD): Checking for Outliers')
plt.xlabel('Income (USD)')
plt.show()
```

# Bivariate Analysis

```python
sns.boxplot(data=df, x='Income (USD)', y='Gender', color='lightblue')
plt.title('Boxplot of Income (USD) by Gender: Checking for Outliers')
plt.xlabel('Income (USD)')
plt.ylabel('Gender')
plt.show()
```

```python
df["Education Level"].unique()
```

```python
sns.boxplot(data=df, x='Income (USD)', y='Education Level', color='lightblue')
plt.title('Boxplot of Income (USD) by Education Level: Checking for Outliers')
plt.xlabel('Income (USD)')
plt.ylabel('Education Level')
plt.show()
```

```python
sns.scatterplot(data=df, x='Age', y='Income (USD)', hue="Gender")
plt.title('Scatter Plot of Age vs Income (USD) by Gender')
plt.xlabel('Age')
plt.ylabel('Income (USD)')
plt.show()
```
