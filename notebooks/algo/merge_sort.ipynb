```python
numbers = [-342, 89, 412, -12, 0, 234, -488, 77, 19, -101, 305, 45, -212, 99, 488, -7, 156, -333, 21, 410, -199, 64, -400, 11, 287, -22, 134, -450, 88, 5, -310, 421, -99, 182, 33, -404, 276, -56, 111, -2, 399, -177, 48, -250, 312, 14, -499, 90, -123, 444, -30, 267, -411, 55, -88, 321, 10, -202, 144, -388, 22, -6, 178, -444, 98, -13, 355, -270, 71, -55, 432, -111, 6, -390, 255, -45, 122, -300, 82, -19, 477, -150, 39, -410, 301, 17, -466, 92, -120, 244, -28, 167, -333, 51, -77, 288, 12, -212, 133, -377]
```

```python
def midpoint(arr):
    return len(arr) // 2
```

```python
def merge(left, right):
    merged = []
    left_index = right_index = 0
    
    while left_index < len(left) and right_index < len(right):
        if left[left_index] < right[right_index]:
            merged.append(left[left_index])
            left_index += 1
        else:
            merged.append(right[right_index])
            right_index += 1
            
    merged.extend(left[left_index:])
    merged.extend(right[right_index:])
    
    return merged
```

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    
    mid = midpoint(arr)
    left_half = merge_sort(arr[:mid])
    right_half = merge_sort(arr[mid:])
    
    return merge(left_half, right_half)
```

```python
sorted_numbers = merge_sort(numbers)
```
