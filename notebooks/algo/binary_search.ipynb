```python
def binary_search(arr, target):
    left, right = 0, len(arr) - 1

    while left <= right:
        mid = left + (right - left) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1
```

```python
no_index = binary_search(sorted_numbers, 500)
no_index
```

```python
print(binary_search(sorted_numbers, -499))
print(binary_search([], 0))
```
