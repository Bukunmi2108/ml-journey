# Implementing complete and partial derivatives in python

```python
def f(x):
    return x ** 2
```

```python
def simple_derivative(f, x, h=1e-8):
    return (f(x) - f(x - h))/h
```

```python
simple_derivative(f, 6)
```

```python
def central_derivative(f, x, h=1e-8):
    return (f(x + h) - f(x - h))/( 2 * h )
```

```python
central_derivative(f, 6)
```

```python
def g(x):
    return (x[0] ** 2) + (3 * x[1] ** 2)
```

```python
def partial_derivative(f, x, i, h=1e-5):
    x_plus = x.copy()
    x_minus = x.copy()
    x_plus[i] += h
    x_minus[i] -= h
    return (f(x_plus) - f(x_minus))/(2 * h)
```

```python
partial_derivative(g, [3, 4], 1)
```

```python
def gradient(f, x, h=1e-5):
    return [partial_derivative(f, x, i, h) for i in range(len(x))]
```

```python
gradient(g, [3, 4])
```

## Analytical Differentiation

```python
import numpy as np
```

```python
x = np.array([1, 2, 3, 4, 5])                          
y = np.array([2, 4, 6, 8, 10])
```

```python
def predict(theta_0, theta_1, x):
    return theta_0 + (theta_1 * x)
```

```python
def mse(theta_0, theta_1, x, y):
    return ((y - predict(theta_0, theta_1, x)) **2).mean()
```

```python
def analytical_derivative(theta_0, theta_1, x, y):
    y_hat = predict(theta_0, theta_1, x)
    errors = y - y_hat
    d_theta_0 = -2 * errors.mean()
    d_theta_1 = -2 * (x * errors).mean()
    return d_theta_0, d_theta_1
```

```python
analytical_derivative(0, 0, x, y)
```

```python
analytical_derivative(0, 2, x, y)
```

```python
J = lambda theta: mse(theta[0], theta[1], x, y)
```

```python
gradient(J, [0.0, 0.0])
```

```python
analytical_derivative(0, 0, x, y)
```

### Autograd

```python
import torch

a = torch.tensor([2., 3., 5.], requires_grad=True)
b = torch.tensor([6., 4., 8.], requires_grad=True)
```

```python
loss = (a * b).mean()
```

```python
loss
```

```python
loss.backward()
```

```python
a.grad
```

```python
b.grad
```
