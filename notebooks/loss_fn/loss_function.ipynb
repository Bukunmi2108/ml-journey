# Loss Functions

### Mean Squared Error

```python
import numpy as np
```

```python
def mean_squared_error(y_true, y_pred):
    """
    Calculate the Mean Squared Error (MSE) between true and predicted values.

    Parameters:
    y_true (numpy array): The true values.
    y_pred (numpy array): The predicted values.

    Returns:
    float: The Mean Squared Error.
    """
    mse = np.mean((y_true - y_pred) ** 2)
    return mse
```

### Binary Cross Entropy

```python
def binary_cross_entropy(y_true, y_pred):
    """
    Calculate the Binary Cross-Entropy loss between true and predicted values.

    Parameters:
    y_true (numpy array): The true binary labels (0 or 1).
    y_pred (numpy array): The predicted probabilities (between 0 and 1).

    Returns:
    float: The Binary Cross-Entropy loss.
    """
    # Clip predictions to avoid log(0)
    y_pred = np.clip(y_pred, 1e-15, 1 - 1e-15)
    bce = -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
    return bce
```

### Categorical Cross Entropy

```python
def categorical_cross_entropy(y_true, y_pred):
    """
    Calculate the Categorical Cross-Entropy loss between true and predicted values.

    Parameters:
    y_true (numpy array): The true one-hot encoded labels.
    y_pred (numpy array): The predicted probabilities for each class.

    Returns:
    float: The Categorical Cross-Entropy loss.
    """
    # Clip predictions to avoid log(0)
    y_pred = np.clip(y_pred, 1e-15, 1 - 1e-15)
    cce = -np.mean(np.sum(y_true * np.log(y_pred), axis=1))
    return cce
```
